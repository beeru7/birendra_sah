```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/iris.csv"
df = pd.read_csv(url)
df
```

```python
df.head()
df.info()
df.describe()
df.columns
```

```python
dfg = df.groupby(['Class'])['Class'].count()
dfg
dfg.plot(kind='bar',
         title='Counts of Iris Classes',
         ylabel='Count',
         xlabel='Class',
         figsize=(16,5))
plt.show()
```

```python
df['SepalLength'].plot(kind='hist')
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
df['SepalLength'].plot(kind='box')
plt.title('Box Plot of Sepal Length')
plt.show()
```

```python
sns.violinplot(y='SepalLength', data=df)
plt.title('Violin Plot of Sepal Length')
plt.show()
```
